```python
# Remove commas from 'Denomination' column
df['Denomination'] = df['Denomination'].str.replace(',', '')

# Convert 'Denomination' column to int
df['Denomination'] = df['Denomination'].astype(int)
```

```python
def extract_first_three_words(s):
    words = s.split()[:2]
    return ' '.join(words)

# create a new column with the first three words of entity because some companies registered like
# FUTURE GAMING AND.... PR, FUTURE GAMING AND.....PVT LTD, FUTURE GAMING AND .... PRIVATE LIMITED
df['Purchaser Name Short'] = df['Purchaser Name'].apply(extract_first_three_words)
```

#Purchaser's Data Analysis

```python
# total unique dates
df['Date of Purchase'].nunique()
```

```python
df.head()
```

```python
#  any missing values in the DataFrame
df.isna().values.any()
```

```python
# total bonds sold on single day (see pdf)

purchase_counts = df.groupby('Date of Purchase')['Denomination'].count().reset_index()
purchase_counts.columns = ['Date of Purchase', 'Total Bond Purchases']
print("total bond sold:", purchase_counts['Total Bond Purchases'].sum())
purchase_counts.head(10)
# purchase_counts.to_csv('total_bond_sold_datewise.csv')
```

```python
# datewise bond amount

purchase_sum = df.groupby('Date of Purchase')['Denomination'].sum().reset_index()
purchase_sum.columns = ['Date of Purchase', 'Total Bond amount']
print("total bond amount purchased:", purchase_sum['Total Bond amount'].sum())
purchase_sum.head(10)
# purchase_counts.to_csv('total_amount_purchased_datewise.csv')
```

```python
# total bond amount purchased
total_amount = purchase_sum['Total Bond amount'].sum()
total_amount_in_crores = total_amount / 10000000
print("total amount of bond purchased:", total_amount_in_crores,"crore")
```

```python
# total amount purchased in a month

# Extract the year and month from the 'Date of Purchase' column
df['Year'] = df['Date of Purchase'].dt.year
df['Month Number'] = df['Date of Purchase'].dt.month

# Calculate the total amount sold for each year and month
total_amount_sold_by_month = df.groupby(['Year', 'Month Number'])['Denomination'].sum().reset_index()
# Rename the columns
total_amount_sold_by_month.columns = ['Year', 'Month Number', 'Total Amount Sold']

print("total amount sold:", total_amount_sold_by_month['Total Amount Sold'].sum())
print("total number of months when bond sold:", total_amount_sold_by_month['Month Number'].count())
print("total amount of bond purchased in 2019:",
      total_amount_sold_by_month[total_amount_sold_by_month['Year'] == 2019]['Total Amount Sold'].sum()
)
total_amount_sold_by_month
```

```python
# total bonds purchased in an year

# Extract the year and month from the 'Date of Purchase' column
df['Year'] = df['Date of Purchase'].dt.year
df['Month'] = df['Date of Purchase'].dt.month

# Calculate the total number of bonds sold for each year and month
total_bonds_sold_by_month = df.groupby(['Year', 'Month'])['Denomination'].count().reset_index()

# Rename the columns
total_bonds_sold_by_month.columns = ['Year', 'Month Number', 'Total Bonds Sold']

print("top five bond selling month and year:\n",
      total_bonds_sold_by_month.sort_values('Total Bonds Sold', ascending=False).head()
)
print("total bond sold since 2019:", total_bonds_sold_by_month['Total Bonds Sold'].sum())
total_bonds_sold_by_month
```

```python
# total amount purchased in an year

# Extract the year from the 'Date of Purchase' column
df['Year'] = df['Date of Purchase'].dt.year

# Calculate the total amount sold for each year
total_amount_sold_by_year = df.groupby('Year')['Denomination'].sum().reset_index()

print("total amount purchased:", total_amount_sold_by_year['Denomination'].sum())
total_amount_sold_by_year
```

# Month wise Bond purchase by each entity

```python
# Extract the year and month from the 'Date of Purchase' column
df['Year'] = df['Date of Purchase'].dt.year
df['Month'] = df['Date of Purchase'].dt.month

# Create a pivot table with 'Purchaser Name Short' as index, years as columns, and months as values
pivot_monthwise_df = df.pivot_table(
    index='Purchaser Name Short',
    columns=['Year', 'Month'],
    values='Denomination',
    aggfunc='sum',
    fill_value=0
)

# Rename the columns to include the year and month
pivot_monthwise_df.columns = [f'{calendar.month_name[month]}_{year}' for year, month in pivot_monthwise_df.columns]
pivot_monthwise_df

# pivot_monthwise_df.to_csv('bond_monthwise_purchase.csv')
```

#Year wise Bond purchase by each entity

```python
# Extract the year from the 'Date of Purchase' column
df['Year'] = df['Date of Purchase'].dt.year

# Create a pivot table with 'Purchaser Name Short' as index and years as columns
pivot_year_df = df.pivot_table(
    index='Purchaser Name Short',
    columns='Year',
    values='Denomination',
    aggfunc='sum',
    fill_value=0
)
pivot_year_df
# pivot_year_df.to_csv('bond_yearwise_purchase.csv')
```
